# analise dos dados

```python
#tabela_salario = filtrar dados 

#tabela_salario_editada = tabela_geral.loc[linha, coluna]
tabela_salario_sf= tabela_geral.loc[tabela_geral["Agency"]=="San Francisco", :]
display(tabela_salario_sf)

tabela_salario_sf.info()
print(tabela_salario_sf.dtypes)
```

```python
tabela_salario_sf['salario'] = pd.to_numeric(tabela_salario_sf['TotalPay'], errors='coerce')
```

```python
tabela_salario_sf.info()
```

```python
#---------------------------
#media ao longo dos anos, dos salarios

#agrupar tudo em uma linha pelo ano
#groupby
#mean() pegar a media

#tabela_sm= tabela_salario_sf.groupby("Year").mean()
tabela_smb = tabela_salario_sf.groupby("Year")["TotalPayBenefits"].mean()
tabela_sm = tabela_salario_sf.groupby("Year")["TotalPay"].mean()
tabela_smid = tabela_salario_sf.groupby("Year")["Id"].mean()

tabela_smb_df= pd.DataFrame(tabela_smb.round(2))
tabela_smid_df= pd.DataFrame(tabela_smid.round(2))
tabela_sm_df= pd.DataFrame(tabela_sm.round(2))

display(tabela_smb_df)
display(tabela_smid_df)
display(tabela_sm_df)
```

# Evolução do salário

```python
#tabela_total= tabela_salario_sf.groupby("Year").sum()
tabela_total= tabela_salario_sf.groupby("Year")[["TotalPay","TotalPayBenefits","Id"]].sum()

display(tabela_total)



tabela_total2= tabela_total[["TotalPay","TotalPayBenefits"]]
display(tabela_total2)

```

```python
#, milhar, e 2f duas casas decimais
#para formatar o numero grandes
def formatar(valor):

    return'R${:,.2f}'.format(valor) 



tabela_total2["TotalPay"]= tabela_total2["TotalPay"].apply(formatar)

tabela_total2["TotalPayBenefits"]= tabela_total2["TotalPayBenefits"].apply(formatar)


display(tabela_total2)
```
