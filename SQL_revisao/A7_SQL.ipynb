# **3. - SELECT TABLE** <p>
A SELECT instrução é usada para selecionar dados de um banco de dados. <p>
*SELECT * FROM table_name* <p>
*SELECT column1, column2, ... FROM table_name*


```python
# Query de consulta - Todas as Colunas
Consulta = Cursor.execute('SELECT * FROM Minha_Tabela').fetchall()

print( Consulta )
```

```python
display(Consulta)
```

```python
# Query de consulta - Colunas Especificas
Consulta = Cursor.execute("""SELECT 
                             Nome,
                             Idade 
                             FROM Minha_Tabela"""                         
                         ).fetchall()

# Loop
for Linha in Consulta:
  print( Linha )
```

```python
Consulta = Cursor.execute("""
    SELECT 
        Nome, 
        Data, 
        SUBSTR(Data, 2, 1) AS teste_data 
    FROM Minha_Tabela
""").fetchall()

for Linha in Consulta:
    print(Linha)
```

```python
# Query de consulta - Colunas Especificas
# No SQL, a contagem começa em 1 (não em 0 como no Python)

Consulta = Cursor.execute("""SELECT 
                             Nome,
                            --- Idade 
                             Data,
                             SUBSTR(Data, 2, 1) AS teste_data 
                             FROM Minha_Tabela"""                         
                         ).fetchall()

# Loop
for Linha in Consulta:
  print( Linha )
```

```python
Consulta = Cursor.execute("""
    SELECT 
        Nome,
        Idade,
        CASE 
            WHEN Idade >= 30 THEN 'S' 
            ELSE 'N' 
        END AS Flag_Trinta
    FROM Minha_Tabela
""").fetchall()

# Visualizando os resultados
for Linha in Consulta:
    print(f"Nome: {Linha[0]} | Idade: {Linha[1]} | Maior ou igual a 30 anos? {Linha[2]}")
```

# **4. - WHERE <p>**
A WHERE cláusula é usada para filtrar registros. <p>
*SELECT column1 FROM table_name WHERE condition;*


**Operadores**
1. '='	Igual
2. '>'	Maior
3. '<'	Menor
4. '>='	Maior ou Igual
5. '<='	Menor ou Igual
6. '<>'	Diferente
7. 'BETWEEN'	Entre
8. 'LIKE'	Procurar por padrão
9. 'IN'	Para especificar vários valores possíveis para uma coluna

```python
# Query usando o Igual "="
Consulta = Cursor.execute(
    '''
    SELECT * FROM Minha_Tabela
    WHERE Nome = 'Odemir'
    '''
).fetchall()

print( Consulta )
```

```python
# Query usando o Igual ">"
Consulta = Cursor.execute(
    '''
    SELECT * FROM Minha_Tabela
    WHERE Idade > 25
    '''
).fetchall()

display( Consulta )
```

```python
# Query usando o Igual "<"
Consulta = Cursor.execute(
    '''
    SELECT * FROM Minha_Tabela
    WHERE Idade < 25
    '''
).fetchall()

display( Consulta )
```

```python
# Query usando o Igual ">="
Consulta = Cursor.execute(
    '''
    SELECT * FROM Minha_Tabela
    WHERE Idade <= 15
    '''
).fetchall()

display( Consulta )
```

```python
# Query usando o diferente "<>", não seja Odemir
Consulta = Cursor.execute(
    '''
    SELECT * FROM Minha_Tabela
    WHERE Nome <> 'Odemir'
    '''
).fetchall()

for Linha in Consulta:
  print( Linha )
```

# **6. - LIKE**
O LIKE operador é usado em uma WHERE cláusula para procurar um padrão especificado em uma coluna. <p>

Existem dois curingas frequentemente usados ​​em conjunto com o LIKEoperador:

 1. O sinal de porcentagem (%) representa zero, um ou vários caracteres
 2. O sinal de sublinhado (_) representa um único caractere

image.png

```python
# Query procurando algo que comece
# no caso comece com O no nome
Consulta = Cursor.execute(
    '''
    SELECT * FROM Minha_Tabela
    WHERE Nome LIKE 'O%'
    '''
).fetchall()

for Linha in Consulta:
  print( Linha )
```

```python
# Query procurando algo que tenha o padrão
# que termina com a letra s
Consulta = Cursor.execute(
    '''
    SELECT * FROM Minha_Tabela
    WHERE Nome LIKE '%s'
    '''
).fetchall()

for Linha in Consulta:
  print( Linha )
```

```python
# Query procurando algo que tenha o padrão
Consulta = Cursor.execute(
    '''
    SELECT * FROM Minha_Tabela
    WHERE Nome LIKE '%ucas%'
    '''
).fetchall()

for Linha in Consulta:
  print( Linha )
```

# **7. - IN**
O IN operador permite especificar vários valores em uma WHEREcláusula. <p>
O IN operador é um atalho para várias ORcondições.

```python
# Query procurando algo com o IN
Consulta = Cursor.execute(
    '''
    SELECT * FROM Minha_Tabela
    WHERE Idade IN (18, 20, 30)
    -- quero valores de 18, 30 e 20 anos, ao invés de usar vários OR
    '''
).fetchall()

for Linha in Consulta:
  print( Linha )
```

```python
# in filtra de forma específica, pesquisa, simplificar a forma de or
```

```python
# Query procurando algo com o IN
Consulta = Cursor.execute(
    '''
    SELECT * FROM Minha_Tabela
    WHERE Nome IN ('Odemir')
    '''
).fetchall()

for Linha in Consulta:
  print( Linha )
```
