```python
# Executar uma consulta para visualizar os dados inseridos
cursor.execute("SELECT * FROM viagens")

# Recuperar todos os resultados
resultados = cursor.fetchall()

# Exibir os resultados
for row in resultados:
    print(row)
```

```python
# Executar a consulta para contar o número de tabelas no banco de dados
cursor.execute("SHOW TABLES")

# Recuperar todos os resultados
tabelas = cursor.fetchall()

# Exibir o número de tabelas
print(f"Quantidade de tabelas no banco de dados: {len(tabelas)}: {tabelas[0]} e {tabelas[1]}")
```
